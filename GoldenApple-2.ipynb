```python
all_links = []
for i in range(1, 56):
    url = 'https://goldapple.ru/makijazh/guby/gubnaja-pomada?p=' + str(i)
    response = requests.get(url, headers={'User-Agent': UserAgent().chrome})
    html = response.content
    soup = BeautifulSoup(html,'html.parser')
    links = soup.find_all('a', {'class' : 'product-item-link js-clamping'})
    for link in links:
        mom = link.get('href')
        all_links.append(mom)
all_links
```
